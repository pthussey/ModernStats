```python
netflix_df = pd.read_csv('datasets/netflix_titles.csv')
netflix_df.head()
```

```python
# Add a release_decade column
netflix_df['release_decade'] = pd.cut(netflix_df.release_year, bins=[1979, 1989, 1999, 2009, 2019], 
                                      labels=['1980s', '1990s', '2000s', '2010s'])
netflix_df.head()
```

```python
# Drop any cases of shows released prior to 1980
netflix_df = netflix_df[netflix_df.release_decade.isna() == False]
netflix_df.reset_index(drop=True, inplace=True)
netflix_df
```

```python
# Drop NR, UR ratings
netflix_df = netflix_df[netflix_df.rating.isin(['NR', 'UR']) == False].copy()
netflix_df.reset_index(drop=True, inplace=True)
netflix_df
```

```python
netflix_df.rating.value_counts()
```

```python
# Group similar ratings
r = ['R', 'TV-MA']
pg_13 = ['TV-14', 'PG-13']
pg = ['PG', 'TV-Y7-FV', 'TV-Y7', 'TV-PG']
g = ['G', 'TV-G', 'TV-Y']

netflix_df.loc[netflix_df.rating.isin(r), 'rating'] = 'R'
netflix_df.loc[netflix_df.rating.isin(pg_13), 'rating'] = 'PG-13'
netflix_df.loc[netflix_df.rating.isin(pg), 'rating'] = 'PG'
netflix_df.loc[netflix_df.rating.isin(g), 'rating'] = 'G'
```

```python
netflix_df.rating.value_counts()
```

```python
# Make rating and release_decade into ordered categorical variables
status_type = CategoricalDtype(categories=['G', 'PG', 'PG-13', 'R'], ordered=True)
netflix_df['rating'] = netflix_df['rating'].astype(status_type)

status_type = CategoricalDtype(categories=['1980s', '1990s', '2000s', '2010s'], ordered=True)
netflix_df['release_decade'] = netflix_df['release_decade'].astype(status_type)
```

```python
# Drop unused columns
netflix_df.drop(labels=['ratinglevel', 'ratingdescription', 'release_year', 'user_rating_score', 'user_rating_size'],
                axis=1, inplace=True)
netflix_df
```

I need to have lists for each rating that contain the proportions for each decade so I can plot the proportions for each rating stacked on top of each other to make the stacked bar chart.

```python
# Create a dictionary that holds a dataframe for each decade
groups = dict(list(netflix_df.groupby('release_decade')))

# Make lists of the rating and release_decade values
decades = sorted(netflix_df.release_decade.unique())
ratings = sorted(netflix_df.rating.unique())

# Build a dictionary that has the rating proportions for each decade
decade_data = {}
for decade in decades:
    decade_data[decade] = groups[decade].rating.value_counts(normalize=True, sort=False)

# Build a rating dictionary that contains lists of proportions for each rating, which is what is needed for plotting
rating_data = {}
for rating in ratings:
    rating_data[rating] = []
    for decade in decades:
        rating_data[rating].append(decade_data[decade][rating])
                    
rating_data
```

```python
# Set up figure (single plot)
fig,ax = plt.subplots()
fig.set_size_inches(16,4)

plt.bar(x=decades, height=rating_data['G'], width=0.6, label='G')
plt.bar(x=decades, height=rating_data['PG'], width=0.6,
        bottom=rating_data['G'], label='PG')
plt.bar(x=decades, height=rating_data['PG-13'], width=0.6,
        bottom=np.array(rating_data['G']) + np.array(rating_data['PG']), label='PG-13')
plt.bar(x=decades, height=rating_data['R'], width=0.6,
        bottom=np.array(rating_data['G']) + np.array(rating_data['PG']) + np.array(rating_data['PG-13']),
        label='R')

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], frameon=True, bbox_to_anchor=(1.0, 1.03), loc='upper left')

ax.set_xlabel('Decade')
ax.set_ylabel('Proportion')
ax.set_title('Netflix Show Ratings by Decade of Release')

plt.show()
```
