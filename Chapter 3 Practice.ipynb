## Case Study: Passwords

The following is my work while following along with the [Chapter 3 case study](https://openintro-ims.netlify.app/data-applications.html#case-study-passwords)

```python
passwords_df = pd.read_csv('datasets/passwords.csv')
passwords_df
```

```python
passwords_df.tail(10)
```

```python
# Remove the empty entries at the bottom
passwords_df = passwords_df[~passwords_df.password.isna()].copy()
passwords_df.tail(10)
```

```python
# Drop the unused columns
passwords_df.drop(columns=['rank_alt', 'font_size'], inplace=True)
passwords_df
```

```python
# Check the columns and their d-types
passwords_df.info(verbose=True)
```

```python
# Create a list of numerical variables, can be passed to the df to reduce to just numerical variables
passwords_numerical_variables = ['rank','value','offline_crack_sec','strength']
passwords_numerical_variables
```

```python
# Create categorical variable df
passwords_categorical_variables = ['password','category','time_unit']
passwords_categorical_variables
```

```python
passwords_df.describe()
```

```python
passwords_df.category.nunique(), passwords_df.category.unique()
```

```python
passwords_df.category.value_counts()
```

```python
passwords_df.isnull().sum()
```

### Initial exploratory analysis through visualization 

```python
# Numerical by histogram
passwords_df[passwords_numerical_variables].hist(bins=15, figsize=(15, 6), layout=(2, 3))
plt.tight_layout()
```

```python
# Numerical by cdf
dsplt.AllNumVarDensityPlot(data=passwords_df, variables=passwords_numerical_variables)
```

```python
# Remove password variable from the list before plotting categorical variables
passwords_categorical_variables.remove('password')
```

```python
# Categorical variables
fig, axes = dsplt.AllCatVarCountPlot(data=passwords_df, variables=passwords_categorical_variables)

for ax in axes:
    ax.tick_params(labelrotation=90)
```

```python
# Make time_unit into an ordinal catergorical variable so that it comes out in order below
status_type = CategoricalDtype(categories=['seconds', 'minutes', 'hours', 'days', 'weeks', 'months', 'years'], ordered=True)
passwords_df['time_unit'] = passwords_df['time_unit'].astype(status_type)
```

```python
# Look at the values grouped by the time_unit variable
passwords_df.groupby(by='time_unit').value.agg(['count', 'min', 'max'])
```

```python
# List top 5 strongest passwords
passwords_df[passwords_df.time_unit == 'years'].value.sort_values(ascending=False).head()
```

```python
# Try a different way that shows all columns
passwords_df[passwords_df.time_unit == 'years'].sort_values(by='value', ascending=False).head()
```

```python
# Look at details of offline_crack_sec
passwords_df.offline_crack_sec.describe()
```

```python
# Plot offline_crack_sec in histogram
passwords_df.offline_crack_sec.plot(kind='hist')
```

```python
# Use matplotlib to make it more granular
fig,ax = plt.subplots()
fig.set_size_inches(8,6)

ax.hist(x=passwords_df.offline_crack_sec.values, bins=range(0,31))

plt.show()
```
